```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
sns.set_context('talk')

import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
```

```python
rews_no_noise_mean = np.array([66.12819643, 71.91923445, 155.145878, 154.9567443])
rews_no_noise_std = np.array([20.77618528, 40.01277323, 41.36339718, 37.79452387])
algorithm_names = ['FD', 'RS', 'CE', 'GA']
```

```python
x = np.arange(len(algorithm_names))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10,8))
rects1 = ax.bar(x, rews_no_noise_mean, width, yerr=rews_no_noise_std)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Reward')
ax.set_xlabel('Algorithm')
ax.set_title('Rewards for No Noise Environments')
ax.set_xticks(x)
ax.set_xticklabels(algorithm_names)
ax.set_ylim([0, 200])

fig.tight_layout()

plt.show()
```

```python
y_means = np.random.rand(50) * 0.5
y_stds = np.random.rand(50) * 0.2
xs = np.arange(50)

plt.plot(xs, y_means)
plt.fill_between(xs, y_means-y_stds, y_means+y_stds, alpha=0.3)
```
