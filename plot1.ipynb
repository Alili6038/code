```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('talk')
sns.set_style('darkgrid')
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
```

```python
N = 4
ind = np.arange(N)  # the x locations for the groups
width = 0.27      # the width of the bars

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)

a = np.array([40.39, 51.13, 52.98, 41.57])
b = np.array([49.93, 54.36, 56.15, 43.97])
c = np.array([27.02, 48.21, 59.99, 58.11])


yvals = [177.71, 148.11, 124.03, 71.11]
rects1 = ax.bar(ind, yvals, width, color='r', yerr=a/2)
zvals = [168, 148.29, 120.12, 69.82]
rects2 = ax.bar(ind+width, zvals, width, color='g', yerr=b/2)
kvals = [192.78, 175.22, 148.41, 89.65]
rects3 = ax.bar(ind+width*2, kvals, width, color='b', yerr=c/2)



ax.set_ylabel('Reward')
ax.set_xlabel('Obs Noise Levels')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('0', '0.25', '0.5', '1') )
ax.legend( (rects1[0], rects2[0], rects3[0]), ('RS', 'CEM', 'GA') )

def autolabel(rects):
    for rect in rects:
        h = rect.get_height()
#         ax.text(rect.get_x()+rect.get_width()/2., 1.05*h, '%d'%int(h),
#                 ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.show()
```
